```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
# visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df =pd.read_csv("student_report_analysis.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed column

```python
df=df.drop(" Unnamed: 0 ",axis=1)
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender")
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

plt.show()
```

#from the above chart we have analysed that:
"the number of femalesin the data is more than the number of males"

```python
gb =df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)

```

```python
plt.figure(figsize=(8,8))
sns.heatmap(gb,annot=True, fmt=".0%")
plt.title("Gender Distribution")
plt.show()
```

#from the above chart we have concluded that:
" the education of the parents have a good impact on their score"

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(gb1,annot=True, fmt=".0%")
plt.title("Relationship between Parent's Educ and Student's Score")
plt.show()
```

from the above chart we have concluded that:
" there is no/negligible impact on the student's score due to their parent's score"

# Distribution of Ethinic Group

```python
group_counts = df['EthnicGroup'].value_counts()
count_list = [group_counts.get("group A", 0), group_counts.get("group B", 0), group_counts.get("group C", 0), group_counts.get("group D", 0), group_counts.get("group E", 0)]
labels = ["group A", "group B", "group C", "group D", "group E"]
plt.figure(figsize=(8, 8))
plt.pie(count_list, labels=labels, autopct="%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```

```python
#To cross check distribution of EthnicGroup
plt.figure(figsize=(5, 5))
ax = sns.countplot(data=df, x='EthnicGroup')

# Annotate the bars with their counts
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

plt.show()
```
